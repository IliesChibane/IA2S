```python
import numpy as np
import matplotlib.pyplot as plt

# Paramètres du système
m = 1.0  # Masse
Kv = 1.0  # Gain proportionnel de vitesse
Kp = 2.0  # Gain proportionnel de position

# Conditions initiales
x0 = 0.0  # Position initiale
v0 = 0.0  # Vitesse initiale

# Temps de simulation
dt = 0.01  # Pas de temps
timesteps = 1500  # Nombre d'itérations (ajusté pour s'assurer que le créneau se termine)

# Initialisation des variables
x = np.zeros(timesteps)
v = np.zeros(timesteps)
f = np.zeros(timesteps)

# Création de la position désirée
xd = np.zeros(timesteps)
xd[:int(10 / dt)] = -1.0
xd[int(10 / dt):] = 1.0

# Boucle de simulation
for i in range(1, timesteps):
    # Calcul de la force de commande
    f[i] = -Kv * v[i-1] + Kp * (xd[i-1] - x[i-1])

    # Calcul de l'accélération (deuxième dérivée de la position)
    a = f[i] / m

    # Intégration numérique pour obtenir la vitesse et la position
    v[i] = v[i-1] + a * dt
    x[i] = x[i-1] + v[i] * dt

# Affichage des résultats
plt.plot(np.arange(0, timesteps * dt, dt), x, label='Position')
plt.plot(np.arange(0, timesteps * dt, dt), xd, label='Position désirée')
plt.plot(np.arange(0, timesteps * dt, dt), v, label='Vitesse')
plt.xlabel('Temps (s)')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Paramètres du système
m = 1.0  # Masse
Kv = 2.0  # Gain proportionnel de vitesse
Kp = 6.0  # Gain proportionnel de position

# Conditions initiales
x0 = 0.0  # Position initiale
v0 = 0.0  # Vitesse initiale

# Temps de simulation
dt = 0.01  # Pas de temps
timesteps = 1500  # Nombre d'itérations (ajusté pour s'assurer que la sinusoïde se termine)

# Initialisation des variables
x = np.zeros(timesteps)
v = np.zeros(timesteps)
f = np.zeros(timesteps)

# Création de la position désirée (sinusoïdale entre +1 et -1)
frequency = 0.1  # Fréquence de la sinusoïde
xd_amplitude = 1.0  # Amplitude de la sinusoïde
xd = xd_amplitude * np.sin(2 * np.pi * frequency * np.arange(0, timesteps * dt, dt))

# Boucle de simulation
for i in range(1, timesteps):
    # Calcul de la force de commande
    f[i] = -Kv * v[i-1] + Kp * (xd[i-1] - x[i-1])

    # Calcul de l'accélération (deuxième dérivée de la position)
    a = f[i] / m

    # Intégration numérique pour obtenir la vitesse et la position
    v[i] = v[i-1] + a * dt
    x[i] = x[i-1] + v[i] * dt

# Affichage des résultats
plt.plot(np.arange(0, timesteps * dt, dt), x, label='Position')
plt.plot(np.arange(0, timesteps * dt, dt), xd, label='Position désirée (sinusoïdale)')
plt.plot(np.arange(0, timesteps * dt, dt), v, label='Vitesse')
plt.xlabel('Temps (s)')
plt.legend()
plt.show()
```
